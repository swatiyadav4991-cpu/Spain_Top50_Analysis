```python
import matplotlib.pyplot as plt
```

```python
top_songs = df['song'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_songs.plot(kind='bar')
plt.title('Top 10 Most Frequent Songs')
plt.xlabel('Songs')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_artists = df['artist'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_artists.plot(kind='bar')
plt.title('Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
album_counts = df['album_type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(album_counts,
        labels=album_counts.index,
        autopct='%1.1f%%')

plt.title('Album Type Distribution')
plt.show()
```

```python
explicit_counts = df['is_explicit'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(explicit_counts,
        labels=['Non-Explicit','Explicit'],
        autopct='%1.1f%%')

plt.title('Explicit vs Non-Explicit Songs')
plt.show()
```

```python
df['popularity'].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['popularity'], bins=20)
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
top_artists = df['artist'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_artists.plot(kind='barh')
plt.title('Top 10 Artists by Playlist Appearances')
plt.xlabel('Appearances')
plt.ylabel('Artist')
plt.show()
```
